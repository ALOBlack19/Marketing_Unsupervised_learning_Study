```python
from sklearn.cluster import KMeans # Unsupervised learning model
from sklearn.metrics import silhouette_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
    Need to calculate the cluster sum of squared distances (WCSS)
        * WCSS in practice:
    -Means the distance between each data point to the centroid of the
     cluster(group of datapoints).
    -WCSS is used to determine the optimal number of clusters(k)
    -The Elbow Method (plot) is a important metric not only for KMeans,
     but also for KNN's.
"""

WCSS = []
k_range = range(1,20) 
"""
    Usually the number of clusters (k) change depending on the
    dataset characteristics, such as:
      * Number of datapoints
      * Quantity of features
      * Data type of the target
      * and so on...
    - That' why the range are from 1 to 20 clusters, to try to find
      the most optimised model parameter:
      * Avoid overfitting
      * Save computational power
"""
for i in k_range:
    Kmeans = KMeans(n_clusters=i, init='k-means++')
    Kmeans.fit(df_scaled)
    WCSS.append(Kmeans.inertia_)
print(WCSS)

plt.figure(figsize=(8, 5))
plt.plot(k_range, WCSS, marker = 'o')
plt.xticks(np.arange(1, 21))
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.show()
```

Observation: Using the __Elbow method__ (ploting) is was possible to verify that the __WCSS value__ (distance between the datapoints and the centroid on the cluster), didn't significatly reduced after k = 2. Based on that visual analysis, k=2 and k=3 can be the most optimized values for this project.

Now with the __Silhouette score curve (plot)__ will be more clear which cluster value is better between k=2 and k=3. 

```python
silhouette_scores = []

for k in range(1, 11):
    Kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)
    cluster_labels = Kmeans.fit_predict(df_scaled)
    if len(set(cluster_labels)) < 2:
        print(f"Skipping k {k} as only {len(set(cluster_labels))} cluster(s) were formed.")
        continue
    score = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='dashed')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal K')
plt.show()
```

Observation: From the silhouette score curve and elbow curve, the most significantly cluster number is k=3. That gives a higher silhouette score.

### __Unsupervised learning model - KMedoid__ 

```python
from sklearn_extra.cluster import KMedoids

plt.rcParams['figure.figsize'] = (6,4)
```

### Fit K-Medoids

```python
# First try with k=5
k = 2
kmedoids = KMedoids(n_clusters=k, metric='euclidean', init='k-medoids++')
labels = kmedoids.fit_predict(df_scaled)

print("Medoid indices", kmedoids.medoid_indices_)
```

### Evaluate clustering quality

```python

silhouette = silhouette_score(df_scaled, labels)
print(f"Silhouette Score: {silhouette:.3f}")
```

```python
print(f"K-Medoids Clustering performed with k={k}")
print("Medoid indices:", kmedoids.medoid_indices_)
silhouette = silhouette_score(df_scaled, labels)
print(f"Silhouette Score (on 28 features): {silhouette:.3f}")
```

### Visualize clusters

```python
df_scaled_array = df_scaled.to_numpy()
# Plot clusters
for lab in np.unique(labels):
    plt.scatter(df_scaled_array[labels == lab, 0], df_scaled_array[labels == lab, 1], label=f'Cluster{lab}')

# Highlight medoids
medoid_points = df_scaled_array[kmedoids.medoid_indices_]
plt.scatter(medoid_points[:,0], medoid_points[:,1], marker='X', s=200, label='Medoids')
plt.xlabel(' x (scaled)')
plt.ylabel('y (scaled)')
plt.title(f'K-Medoids Clusters (k={k})')
plt.legend()
plt.show()

```

After plotting the clustered data it was possible to visualize a case of high number of dimensions caused by the amount of features present on the original data, and even higher, after the encoding step.

This bad visualization, with "scrambled" and overlapping view, is due to the trying of plotting 28 features dimensions in only 2 dimension graph.

#### Based on the fact above, I'll perform PCA to make __dimensionality reduction__

### __PCA - Principal Component Analysis__

 - Doing PCA, the mode will be able to analyse the dataset and select the most important features, and then reduce the amount of dimensions for better optimiziation, visualization and resource saving (time and energy) to run the model.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

df_pca = pca.fit_transform(df_scaled_array)
print(f"\nExplained variance by 2 principal components: {pca.explained_variance_ratio_.sum():.2f}")
print(f"PC1 explains: {pca.explained_variance_ratio_[0]*100:.1f}%")
print(f"PC2 explains: {pca.explained_variance_ratio_[1]*100:.1f}%")
```

### Fit K-Medoids -- __PCA__

```python
# First try with k=5
k = 2
kmedoids = KMedoids(n_clusters=k, metric='euclidean', init='k-medoids++')
labels = kmedoids.fit_predict(df_pca)

print("Medoid indices", kmedoids.medoid_indices_)
```

### Evaluate clustering quality __PCA__

```python

silhouette = silhouette_score(df_pca, labels)
print(f"Silhouette Score: {silhouette:.3f}")
```

### Visualize clusters __PCA__

```python
# Plot clusters
for lab in np.unique(labels):
    plt.scatter(df_pca[labels == lab, 0], df_pca[labels == lab, 1], label=f'Cluster{lab}')

# Highlight medoids
medoid_points = df_pca[kmedoids.medoid_indices_]
plt.scatter(medoid_points[:,0], medoid_points[:,1], marker='X', s=200, label='Medoids')
plt.xlabel(' x (scaled)')
plt.ylabel('y (scaled)')
plt.title(f'K-Medoids Clusters (k={k})')
plt.legend()
plt.show()

```

### __Unsupervised learning model - DBSCAN__ 

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

plt.rcParams['figure.figsize'] = [6,4]
```

__Density Based Spatial Clustering of Applications With Noise__ is used for discovering the clusters arbitary shape and handling noise.

The DBCSAN cluster shape be arbitrary means that the cluster format does not conform to a predefined geometric form. In another words __different clustering algorithms are desingned to find different types of shapes__. This is important because in real world data not all clusters have a predefined geometric forms with distances patterns and density with a centralized point such as __centroids for KMeans__ and __Medoids for KMedoids__.

Arbitrary shapes helps marking outliers points that lie alone in low density and unusual format cluster regions (more sparse clusters points with alongaded or S-shape).

Fitting __DBSCAN__ at this dataset to check the theory behind.

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(df_scaled)

# First, calculate the total count of unique labels
total_unique_labels = len(set(labels))

# Then, apply the conditional subtraction based on whether -1 is present
if -1 in labels:
    n_clusters = total_unique_labels - 1
else:
    n_clusters = total_unique_labels

n_noise = list(labels).count(-1)

print(f"Clusters found: {n_clusters}, noise points: {n_noise}")
```

```python
# Silhouette only if >1 cluster and some nonâ€‘noise points
if n_clusters > 1:
    mask = labels != -1
    sil = silhouette_score(df_scaled[mask], labels[mask])
    print(f"Silhouette Score (excluding noise): {sil:.3f}")
else:
    print("Silhouette Score not defined for a single cluster.")
```

```python
unique_labels = set(labels)
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_labels)))
for (lab, col) in zip(unique_labels, colors):
    class_member_mask = labels == lab
    if lab == -1:
        plt.scatter(df_scaled_array[class_member_mask,0], df_scaled_array[class_member_mask,1],
                    c='k', marker='x', s=30, label='Noise')
    else:
        plt.scatter(df_scaled_array[class_member_mask,0], df_scaled_array[class_member_mask,1],
                    c=[col], s=20, label=f'Cluster {lab}')
plt.legend()
plt.title('DBSCAN result')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.show()
```
